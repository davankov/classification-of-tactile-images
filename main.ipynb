```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/19-30-03.csv")
coords = [(.3, .9), (.5, .9), (.7, .9),
          (.2, .7), (.4, .7), (.6, .7), (.8, .7),
          (.1, .5), (.3, .5),(.5, .5), (.7, .5), (.9, .5),
          (.2, .3), (.4, .3), (.6, .3), (.8, .3),
          (.3, .1), (.5, .1), (.7, .1)]

colors = {0: "lightgreen", 1: "palegreen", 2: "mediumspringgreen", 3: "springgreen",
          4: "chartreuse", 5: "lime", 6: "lime", 7: "yellow", 8: "yellow",
          9: "gold", 10: "gold", 11: "orange", 12: "orange", 13: "darkorange",
          14: "darkorange", 15: "tomato", 16: "tomato", 17: "orangered",
          18: "orangered", 19: "red", 20: "red", 21: "red", 22: "black", 23: "black", 24: "black", 25: "black"}
plt.plot(data)
plt.show()
for j in range(data.shape[0]):
    strength = data.iloc[j]
    for i in range(len(coords)):
        c1 = plt.Circle(coords[i], .05, color=colors[strength[i] // 10])
        plt.gcf().gca().add_artist(c1)
    plt.show()
data
```
